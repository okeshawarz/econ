# New Interpretation Demonstration 
The NI introduces two new interpretations of the LTV: the value of money and the value of labor-power. This allows for the equivalence of value and monetary quantities.

The *Value of Money* is defined as the ratio of total productive labor (in hours) and the total money value added (measured in units of dollars). In other words, this is the ratio of productive labor to money value added, or labor hours per dollar. 

The *Value of Labor-power* is the product of the nominal wage rate and the value of money, i.e. the social labor-time equivalent of the nominal wage. In other *other* words, this is the wage in terms of *value* rather than in the prices of subsistence/wage goods.

```julia
using LinearAlgebra, LaTeXStrings, Latexify

#### Given Quantities
# A = input output matrix
# l = labour input vector
# w = nominal wage rate
# v = value of labour power
# Q = gross output vector

A = [186/450 54/21 30/60;
     12/450 6/21 3/60;
     9/450 6/21 15/60]
l = [18/450, 12/21, 30/60]
w = 1
v = 0.515
y = [180, 0, 30]

#### Computed Quantities
# Gross Output Vector
Q = inv(I-A) * y
eigmax = maximum(eigvals(A))

#Maximal rate of profit
R = (1 / eigmax) - 1

#Define univariate function rate of profit
function f(r2)
     D2 = (I - A) * Q
     l2 = l
     A2 = A
     L2 = (l * Q) / v
     C2 = (I-(1+r2)\A2) ## Look at this if things go wrong
     E2 = C2 * D2
     B2 = (1 + r2) * l2 * E2
     i2 = B2 - L2
     return(i2)
end
using Roots
r = find_zero(f, (0, R-0.01))

```
